```python
import pandas as pd
import numpy as np
import matplotlib as mtpl
```

```python
credits = pd.read_csv('movie_recomndation/tmdb_5000_credits.csv')
movie  = pd.read_csv('movie_recomndation/tmdb_5000_movies.csv')
```

```python
df = df[['genres','id','title','overview','keywords','cast','crew']]
```

```python
df.dropna(inplace=True)
```

```python
import ast
```

```python
def convert(obj):
    value=[]
    for i in ast.literal_eval(obj):
        value.append(i['name'])
    return value
```

```python
df['genres']=df['genres'].apply(convert)
```

```python
df['genres'][0]
```

```python
df['keywords']=df['keywords'].apply(convert)
```

```python
df.head()
```

```python
df['cast'][0]
```

```python
def convert_3(obj):
    value=[]
    count=0
    for i in ast.literal_eval(obj):
        if count != 3:
            value.append(i['name'])
            count+=1
        else:
            break
    return value
```

```python
df['cast'] = df['cast'].apply(convert_3)
```

```python
df.head()
```

```python
def convert_director(obj):
    value=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            value.append(i['name'])
            break
        
    return value
```

```python
df['crew'] = df['crew'].apply(convert_director)
```

```python
df.head()
```

```python
df['overview']=df['overview'].apply(lambda x:x.split())
```

```python
df.head()
```

```python
df['genres']=df['genres'].apply(lambda x:[i.replace(" ","")for i in x])
df['keywords']=df['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
df['cast']=df['cast'].apply(lambda x:[i.replace(" ","")for i in x])
df['crew']=df['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
df.head()
```

```python
df['tags'] = df['overview']+df['genres']+df['cast']+df['crew']+df['keywords']
```

```python
df = df[['id','title','tags']]
```

```python
df.head()
```

```python
df['tags']=df['tags'].apply(lambda x:" ".join(x))
```

```python
df.head()
```

```python
df['tags'] = df['tags'].apply(lambda x:x.lower())
```

```python
df.head()
```

```python
df['tags'].nunique()
```

```python
from nltk.stem.porter import PorterStemmer
# stemming the tags data
ps = PorterStemmer()

def stem(text):
    val=[]
    for i in text.split():
        val.append(ps.stem(i))
    return " ".join(val)
```

```python
df['tags']=df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=4000,stop_words='english')
```

```python
vector = cv.fit_transform(df['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
# taking cosine dist for similarity
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
df[df['title']=='Avatar'].index[0]
```

```python
# we cant just use sorted() because we are comparing similarity against movie index and in sorting we loose the index
def recommend(movie):
    movie_index = df[df['title']==movie].index[0]
    dist = similarity[movie_index]
    movie_list = sorted(list(enumerate(dist)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movie_list:
        print(df.iloc[i[0]].title)
```

```python
recommend('Titanic')
```
